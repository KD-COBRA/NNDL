```python
# Task-1
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-10, 10, 100)
y_step = step_function(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_step, label='Step Function')
plt.title('Step Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_sigmoid = sigmoid(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_sigmoid, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()

def bipolar_sigmoid(x):
    return 2 / (1 + np.exp(-x)) - 1

y_bipolar = bipolar_sigmoid(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_bipolar, label='Bipolar Sigmoid Function')
plt.title('Bipolar Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()

def tanh(x):
    return np.tanh(x)

y_tanh = tanh(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_tanh, label='Tanh Function')
plt.title('Tanh Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()

def relu(x):
    return np.maximum(0, x)

y_relu = relu(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_relu, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()
```

```python
# Task-2
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# XOR Dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Function to create and compile a model
def create_model(activation_function):
    model = Sequential()
    model.add(Dense(2, input_dim=2, activation=activation_function))
    model.add(Dense(1, activation='sigmoid'))  # Output layer
    model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
    return model

# Evaluate each activation function
activation_functions = ['sigmoid', 'tanh', 'relu']
results = {}

for activation in activation_functions:
    model = create_model(activation)
    model.fit(X, y, epochs=1000, verbose=0)
    loss, accuracy = model.evaluate(X, y, verbose=0)
    results[activation] = accuracy

print("Performance of different activation functions on XOR problem:")
for activation, accuracy in results.items():
    print(f"{activation}: {accuracy:.2f}")
```
